```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# import the data
data = pd.read_csv("/content/train.csv")
data
```

```python
x_train = data[["Pell_grant_rate"]].values
y_train = data["Completion_rate"]
print(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
# Use Linear Regression to get the model based on the train dataset and get the score
model = LinearRegression().fit(x_train, y_train)
model_score = model.score(x_train, y_train)
print(model_score)
```

```python
# Use the test dataset to get the predicted result
test = pd.read_csv("/content/x_test.csv")
x_test = test[["Pell_grant_rate"]].values
y_test = model.predict(x_test)
print(y_test)
```

```python
submission = pd.DataFrame.from_dict({'Completion_rate': y_test})
submission
```

```python
submission.to_csv('submission.csv', index = True, index_label = "id") 
```
